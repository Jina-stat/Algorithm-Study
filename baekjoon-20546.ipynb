```python
# 백준 20546

import sys
data = sys.stdin.read().splitlines()

# s=open("baekjoon-20546-data.txt","rt")
# data = s.read().splitlines()

money1 = int(data[0])
money2 = int(data[0])
stock_list = list(map(int, data[1].split(" ")))

rate_n = 0
rate = 0

stock1 = 0
stock2 = 0

for idx, i in enumerate(stock_list):
    # 준현
    stock1 += money1 // i
    money1 = money1 % i

    # 성민
    if idx >= 1:
        
        # # 두번째 날부터 오늘 기준 증가, 감소정도 계산
        rate = i - stock_list[idx - 1]
        if rate_n >= 0 and rate > 0: # 증가 or 유지 -> 증가
            rate_n += 1

        elif rate_n >= 0 and rate < 0: # 증가 or 유지 -> 감소
            rate_n = -1

        elif rate_n <=0 and rate > 0: # 감소 or 유지 -> 증가
            rate_n = 1

        elif rate_n <=0 and rate < 0: # 감소 or 유지 -> 감소
            rate_n -= 1
            
        elif rate == 0: # 유지
            rate_n = 0 
        
        # print(i, money2, stock2, rate_n)
        if rate_n >= 3: #세번 연속 증가 했으면, 모두 매도 
            money2 += stock2 * i
            stock2 = 0
            
            
        elif rate_n <= -3: #세번 연속 감소 했으면, 모두 매수 
            stock2 += money2 // i
            money2 = money2 % i
            


value1 = money1 + stock_list[-1] * stock1
value2 = money2 + stock_list[-1] * stock2

if value1 > value2:
    print("BNP")
elif value1 < value2:
    print("TIMING")
else:
    print("SAMESAME")


```
